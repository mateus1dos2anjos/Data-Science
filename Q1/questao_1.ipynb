```python
import glob
import os
import pandas as pd
from collections import Counter
```

```python
all_files = glob.glob(os.path.join("data", "*.xlsx"))  
dataframe = pd.concat((pd.read_excel(f) for f in all_files), ignore_index=True)
```

```python
dataframe
```

```python
content = dataframe['Conteúdo']
content
```

```python
content = content.dropna()
```

```python
content.isnull().values.any()
```

```python
words_set = Counter()
content.str.lower().str.split().apply(words_set.update)
```
